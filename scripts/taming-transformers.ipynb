## Load example data

Load an example segmentation and visualize.

```python
from PIL import Image
import numpy as np
segmentation_path = f"{project_path}/data/sflckr_segmentations/norway/25735082181_999927fe5a_b.png" # Its a grey scale image with shape (H, W)
segmentation = Image.open(segmentation_path)
segmentation = np.array(segmentation)
segmentation = np.eye(182)[segmentation] # For each pixel in H*W pixels, convert it to a C=182 one-hot vector. So that the shape of this tensor is now (H, W , C)


# Convert the one-hot encoded NumPy array to a PyTorch tensor. 
# The transpose(2,0,1) changes the shape from (H, W, C) to (C, H, W), and [None] adds a batch dimension, resulting in a tensor of shape (1, C, H, W). 
# Finally, it is converted to float32 and moved to the device (CPU or GPU) used by the model.
segmentation = torch.tensor(segmentation.transpose(2,0,1)[None]).to(dtype=torch.float32, device=model.device)
```

Visualize

```python
def show_segmentation(s):
  s = s.detach().cpu().numpy().transpose(0,2,3,1)[0,:,:,None,:] # Move tensor to CPU, convert to NumPy array and rearrange axes
  colorize = np.random.RandomState(1).randn(1,1,s.shape[-1],3) # Generate random colors for each class
  colorize = colorize / colorize.sum(axis=2, keepdims=True)  # Normalize the colors
  s = s@colorize # Apply colors to the segmentation map
  s = s[...,0,:] # Remove the singleton dimension
  s = ((s+1.0)*127.5).clip(0,255).astype(np.uint8) # Scale to 0-255 and convert to uint8
  s = Image.fromarray(s)# Convert to a PIL image
  display(s)

show_segmentation(segmentation)
```

Our model also employs a VQGAN for the conditioning information, i.e. the segmentation in this example. Let's autoencode the segmentation map. Encoding returns both the quantized code and its representation in terms of indices of a learned codebook.

```python
c_code, c_indices = model.encode_to_c(segmentation) # segmentation has shape (B=1, C=128, H, W)
print("c_code", c_code.shape, c_code.dtype) # The quanzited embedding z_q has shape (B=1, C=256 which is the dimension number of each code e_k in the codebook, h, w), consisting of h*w codes
print("c_indices", c_indices.shape, c_indices.dtype) # Since z_q is composed of h*w codes, the amount of corresponding indices is h*w
assert c_code.shape[2]*c_code.shape[3] == c_indices.shape[0] # Assert that len(indices)==h*w
segmentation_rec = model.cond_stage_model.decode(c_code)
show_segmentation(torch.softmax(segmentation_rec, dim=1)) # Since `segmentation_rec` has shape (B, C, H, W). We apply softmax to axis C.
```

Let's sample indices corresponding to codes from the image VQGAN given the segmentation code. We init randomly and take a look.

```python
def show_image(s):
  s = s.detach().cpu().numpy().transpose(0,2,3,1)[0]
  s = ((s+1.0)*127.5).clip(0,255).astype(np.uint8)
  s = Image.fromarray(s)
  display(s)

codebook_size = config.model.params.first_stage_config.params.embed_dim # `codebook_size` is the dimension number of each code e_k in the codebook.
z_indices_shape = c_indices.shape # h*w
z_code_shape = c_code.shape # the shape of z_q, (B=1, C=256 which is the dimension number of each code e_k in the codebook, h, w)

# Randimly generate indices to sample from the VQ-GAN.
# low=0, high=`z_indices`=256. So each index ranges from 0 to 256-1. There are `len(z_indices_shape)`=h*w indices in total.
z_indices = torch.randint(codebook_size, z_indices_shape, device=model.device)
x_sample = model.decode_to_img(z_indices, z_code_shape)
show_image(x_sample)
```

## Sample an image

We use the transformer in a sliding window manner to sample all code entries sequentially. The code below assumes a window size of $16\times 16$.

```python
from IPython.display import clear_output
import time

'''
This code performs pixel-by-pixel sampling to generate an image based on an initial set of indices and a conditional input. It does this by:

1. Iterating over the spatial dimensions of the target tensor.
2. Extracting local patches around each pixel.
3. Running a transformer model to get logits for the current pixel.
4. Sampling from the logits to update the index tensor.
5. Periodically decoding and displaying the generated image.
'''
idx = z_indices # (h*w)
idx = idx.reshape(z_code_shape[0],z_code_shape[2],z_code_shape[3]) # (B=1, h, w)

cidx = c_indices 
cidx = cidx.reshape(c_code.shape[0],c_code.shape[2],c_code.shape[3]) # (B=1, h, w)

temperature = 1.0
top_k = 100
update_every = 50

start_t = time.time()

# Loop Over Spatial Dimensions (h, w) with index (i, j)
# Local Indices: local_i and local_j determine the position within a 16x16 patch. This ensures the patch is correctly aligned within the bounds of the tensor.
for i in range(0, z_code_shape[2]-0):
  if i <= 8:
    local_i = i
  elif z_code_shape[2]-i < 8:
    local_i = 16-(z_code_shape[2]-i)
  else:
    local_i = 8
  for j in range(0,z_code_shape[3]-0):
    if j <= 8:
      local_j = j
    elif z_code_shape[3]-j < 8:
      local_j = 16-(z_code_shape[3]-j)
    else:
      local_j = 8

    # i_start, i_end, j_start, j_end define the indices for a 16x16 patch centered around (i, j).
    i_start = i-local_i
    i_end = i_start+16
    j_start = j-local_j
    j_end = j_start+16
    
    patch = idx[:,i_start:i_end,j_start:j_end] # patch extracts a 16x16 region from idx.
    patch = patch.reshape(patch.shape[0],-1) # patch and cpatch are reshaped into vectors (flattened).
    cpatch = cidx[:, i_start:i_end, j_start:j_end]
    cpatch = cpatch.reshape(cpatch.shape[0], -1)
    patch = torch.cat((cpatch, patch), dim=1)
    logits,_ = model.transformer(patch[:,:-1]) # runs the transformer model on the concatenated patch (excluding the last element).
    logits = logits[:, -256:, :]
    logits = logits.reshape(z_code_shape[0],16,16,-1) # logits are reshaped back to match the spatial dimensions and select the relevant part.


    logits = logits[:,local_i,local_j,:]

    logits = logits/temperature

    if top_k is not None:
      logits = model.top_k_logits(logits, top_k)

    probs = torch.nn.functional.softmax(logits, dim=-1)
    idx[:,i,j] = torch.multinomial(probs, num_samples=1) # samples an index from the probability distribution for the current position (i, j).

    step = i*z_code_shape[3]+j
    if step%update_every==0 or step==z_code_shape[2]*z_code_shape[3]-1: # Every update_every steps, or at the last step, the current image is decoded and displayed.
      x_sample = model.decode_to_img(idx, z_code_shape)
      clear_output()
      print(f"Time: {time.time() - start_t} seconds")
      print(f"Step: ({i},{j}) | Local: ({local_i},{local_j}) | Crop: ({i_start}:{i_end},{j_start}:{j_end})")
      show_image(x_sample)
```
